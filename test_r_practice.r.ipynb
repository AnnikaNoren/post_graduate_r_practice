## creating an object

Simplest object class is a vector, a collection of values of the same class

Vector plus date is a time series indicated by prefix "ts"

numeric vector contains numeric data with decimals

integer vector contains only integers

R is obsessed with vectors

arrows work in both directions as long as it points to the object name

to see a list of objects created in console, use ls() or objects()

remove an object from environment use rm("name")



## Working with Vectors

## Functions
sequence: seq(from=, to=, by = )

paste: concatenate vectors after converting to strings

check class: class(paste(1:4), result is character, used for numbered characters for plot legends

rep: replicate elements of vectors and lists, each and times arguments


working with index positions: which(x == 10) double equal sign and logical operator
